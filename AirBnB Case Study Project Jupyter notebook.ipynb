```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head(40)
# air_bnb.shape
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.neighbourhood_group.value_counts()

#  5 groups available and Manhattan shows up the most with 21,661 total
```

```python
# Which hosts are the busiest and based on their reviews?
top_host_numbers = air_bnb[['host_name','number_of_reviews']].sort_values(by=['number_of_reviews'], ascending=False).head(50)
top_host_numbers_grouped_by_host_name = top_host_numbers.groupby("host_name")
added = top_host_numbers_grouped_by_host_name.sum()
sorted_reviews = added.sort_values(by=['number_of_reviews'], ascending=False).reset_index()
sorted_reviews

# Maya, Jj, Danielle and Dona all have over 1000 reviews with Maya being the highest at 2,273 reviews. 
```

```python
#Which neighorhood group has the highest average price?
highest_average_price = air_bnb.groupby(["neighbourhood_group"]).mean().reset_index()
highest_average_price

# Manhattan had the highest average price at $196.87 dollars. 
```

```python
# Which neighbor hood group has the highest total price?
highest_price_sum = air_bnb.groupby(["neighbourhood_group"]).sum().reset_index()
print(highest_price_sum[["neighbourhood_group",'price']].sort_values(by = "price", ascending = False))

# Manhattan had the highest total price. 
```

```python
#Which top 5 hosts have the highest total price?

top_host_price = air_bnb[['host_name','price']].sort_values(by=['price'], ascending=False).head(50)
top_host_price_w_name = top_host_price.groupby("price")
top_host_price_w_name.first().reset_index().sort_values(by=['price'], ascending=False).head(50)

# Katherine, Matt, Rum, Jessica and Sally are top 5 based on price. 
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?


air_bnb["Host Check"] = ' '

def bookedHost(row):
    superHost = row["Host Check"]
    reviews = row["number_of_reviews"]
    availibility = row["availability_365"]
    if 100 <= reviews and 0 == availibility:
        return "Super Host"
    else:
        return "Reg Host"
    
air_bnb["Host Check"] = air_bnb.apply(bookedHost, axis = 1)

super_host_only = air_bnb[air_bnb["Host Check"]== "Super Host"]
simple_super_host_only= super_host_only[['host_name','number_of_reviews','availability_365','Host Check']].head(100)

grouped_super_host= simple_super_host_only.groupby("host_name")
added_up = grouped_super_host.sum()

highest_reviews_no_availability = added_up.sort_values(by=['number_of_reviews'], ascending=False).reset_index()
print(highest_reviews_no_availability)
```
